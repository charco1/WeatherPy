```python
# Range of latitudes and longitudes
latitude_range = (-90, 90)
longitude_range = (-180, 180)
latitude_longitude = []
cities = []

latitudes = np.random.uniform(low=-90.000, high=90.000, size=1500) #randon list of latitudes
longitudes = np.random.uniform(low=-180.000, high=180.000, size=1500) #random list of longitudes
latitudes_longitudes = zip(latitudes, longitudes)

# For latitude and longitude 
for latitude_longitude in latitudes_longitudes:
    city = citipy.nearest_city(latitude_longitude[0], latitude_longitude[1]).city_name
    
    if city not in cities:
       cities.append(city)

# Print the city count to verify number of cities
len(cities)
```

```python
# Calling URL for Weather API
url = "http://api.openweathermap.org/data/2.5/weather?" 
units = "imperial"

api_keys = "4859ae2d55a28956a18b356fa5785d7b"

search_url = url + "appid=" + api_keys + "&units=" + units + "&q="
```

```python
# Loop through the list of cities and perform request for data on each
search_data = []
cityList = []
countryList = []
randCity = []
randCountry = []

loopcnt = 0
# get = () 
import json


#city = citipy.nearest_city(latitudes, longitudes)

#Loop 
while len(randCity) < 500 :
    # Loop counter to track number of loops made
    loopcnt +=1
# Randomly generate latitude
    latitude_pos = random.randint(0,2) # generate random numbers (start, end)
    base_latitude = random.randint(0,91)
    dec_latitude = random.random()/100
    latitude = base_latitude + dec_latitude
    if latitude_pos == 1:
        latitude = latitude * -1
    print(f"The latitude is: {latitude}") 
    
    # Randomly generate longitude
    longitude_pos = random.randint(0,2)
    base_longitude = random.randint(0,181)
    dec_longitude = random.random()/100
    longitude = base_longitude + dec_longitude
    if longitude_pos == 1:
        longitude = longitude * -1
    print(f"The longitude is: {longitude}")
    
    # Call citipy file to get nearest city based on random lat and lng
    city = citipy.nearest_city(latitude, longitude)
    
    foundCity = False
    for j in range(len(randCity)):
        if city.city_name == randCity[j]:
            foundCity = True
    if foundCity==False  :   
        randCity.append(city.city_name)
        randCountry.append(city.country_code)
    
searchCnt = 0
print("Begin API Call to OpenWeather Database ")
print("---------------------------------------")
for city in randCity:
    searchCnt += 1
    print(f"Processing record {searchCnt} for the city: {city}")   
    response = requests.get(search_url + city).json()
    searchCityID = response.get("id")
    if response.get("id"):
        print(f"     Record found for city: {city}    city id: {searchCityID}")
        search_data.append(response)
        cityList.append(city)
        countryList.append(randCountry[searchCnt-1])
    else:
        print(f"     No weather record found for city: {city}")
          
print("---------------------------------------")
print("Processing Complete")
```

```python
# Use list comprehension to gather necessary data series

latitude_data = [data.get("coord").get("lat") for data in search_data]
longitude_data = [data.get("coord").get("lon") for data in search_data]
temperature_data = [data.get("main").get("temp") for data in search_data]
humidity_data = [data.get("main").get("humidity") for data in search_data]
cloudiness_data =[data.get("clouds").get("all") for data in search_data]
wind_data = [data.get("wind").get("speed") for data in search_data]
```
